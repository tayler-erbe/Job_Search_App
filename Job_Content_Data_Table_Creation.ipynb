```python
import openpyxl
from openpyxl import Workbook
import os

# Define the input file and output directory
input_file = r"C:\Users\terbe\Desktop\Job Classification - Cosine Similarity & GenAI Experimentation\Class Index Export 03.28.2024 (1).xlsx"
output_directory = r"C:\Users\terbe\Downloads"

# Load the workbook
workbook = openpyxl.load_workbook(input_file)

# Ensure the output directory exists
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Save each sheet as a separate file
for sheet_name in workbook.sheetnames:
    # Load the sheet
    sheet = workbook[sheet_name]
    
    # Create a new workbook for the sheet
    new_workbook = Workbook()
    new_sheet = new_workbook.active
    new_sheet.title = sheet_name

    # Copy data from the original sheet to the new workbook
    for row in sheet.iter_rows(values_only=True):
        new_sheet.append(row)
    
    # Save the new workbook with the sheet name as the file name
    output_file = os.path.join(output_directory, f"{sheet_name}.xlsx")
    new_workbook.save(output_file)
    print(f"Saved: {output_file}")
```

```python
import pandas as pd
import os

# List of file paths
file_paths = [
    r"C:\Users\terbe\Downloads\Root Index.xlsx",
    r"C:\Users\terbe\Downloads\Promotional Lines.xlsx",
    r"C:\Users\terbe\Downloads\Job Duties Fixed.xlsx",
    r"C:\Users\terbe\Downloads\Job Duties.xlsx",
    r"C:\Users\terbe\Downloads\E-Test Categories.xlsx",
    r"C:\Users\terbe\Downloads\KSA.xlsx",
    r"C:\Users\terbe\Downloads\Levels of Work.xlsx",
    r"C:\Users\terbe\Downloads\MAQ.xlsx",
    r"C:\Users\terbe\Downloads\MAQ HTML Only.xlsx",
    r"C:\Users\terbe\Downloads\MAQ CSS.xlsx"
]

# Dictionary to store dataframes
dataframes = {}

# Check file existence and load
for file_path in file_paths:
    if os.path.exists(file_path):
        sheet_name = file_path.split("\\")[-1].replace(".xlsx", "")  # Extract file name as key
        dataframes[sheet_name] = pd.read_excel(file_path)
        print(f"Loaded: {file_path}")
    else:
        print(f"File not found: {file_path}")

# Display the names of the successfully loaded dataframes
print("\nLoaded DataFrames:", dataframes.keys())
```

```python
ksa_df = dataframes.get('KSA')

# Update the grouping to include both "Class Code" and "Class Title"
if "Class Code" in ksa_df.columns and "Class Title" in ksa_df.columns:
    # Remove decimals, round to int, and convert data type for "Class Code"
    ksa_df["Class Code"] = ksa_df["Class Code"].fillna(0).astype(float).round(0).astype(int)

    # Group by "Class Code" and "Class Title", and concatenate "Levels Of Work" strings
    ksa_df = ksa_df.groupby(["Class Code", "Class Title"])['KSA'].apply(lambda x: ' '.join(x.astype(str))).reset_index()


else:
    print("The 'Class Code' column is not found in the DataFrame.")

ksa_df
```

```python
# Retrieve the 'Levels of Work' DataFrame
levels_of_work_df = dataframes.get('Levels of Work')

# Update the grouping to include both "Class Code" and "Class Title"
if "Class Code" in levels_of_work_df.columns and "Class Title" in levels_of_work_df.columns:
    # Remove decimals, round to int, and convert data type for "Class Code"
    levels_of_work_df["Class Code"] = levels_of_work_df["Class Code"].fillna(0).astype(float).round(0).astype(int)

    # Group by "Class Code" and "Class Title", and concatenate "Levels Of Work" strings
    levels_of_work_df = levels_of_work_df.groupby(["Class Code", "Class Title"])["Levels Of Work"].apply(lambda x: ' '.join(x.astype(str))).reset_index()


else:
    print("The 'Class Code' column is not found in the DataFrame.")

levels_of_work_df
```

```python
maq_df = dataframes.get('MAQ')
# Fill NaN values in "Class Code" and "Class Title" columns with the previous non-empty value
maq_df["Class Code"] = maq_df["Class Code"].fillna(method="ffill")
maq_df["Class Title"] = maq_df["Class Title"].fillna(method="ffill")


# Update the grouping to include both "Class Code" and "Class Title"
if "Class Code" in maq_df.columns and "Class Title" in maq_df.columns:
    # Remove decimals, round to int, and convert data type for "Class Code"
    maq_df["Class Code"] = maq_df["Class Code"].fillna(0).astype(float).round(0).astype(int)

    # Group by "Class Code" and "Class Title", and concatenate "Levels Of Work" strings
    maq_df = maq_df.groupby(["Class Code", "Class Title"])['MAQ'].apply(lambda x: ' '.join(x.astype(str))).reset_index()


else:
    print("The 'Class Code' column is not found in the DataFrame.")

maq_df


```

```python
import pandas as pd

# Assuming all DataFrames have been initialized: root_index_df, job_duties_fixed_df, ksa_df, levels_of_work_df, maq_df

# Step 1: Left join root_index_df with job_duties_fixed_df on ["Class Code", "Class Title"]
merged_table = root_index_df.merge(
    job_duties_fixed_df,
    on=["Class Code", "Class Title"],
    how="left"
)

# Step 2: Left join merged_table with ksa_df on ["Class Code", "Class Title"]
merged_table = merged_table.merge(
    ksa_df,
    on=["Class Code", "Class Title"],
    how="left"
)

# Fill empty values in ksa_df["KSA"] with an empty string
if "KSA" in merged_table.columns:
    merged_table["KSA"] = merged_table["KSA"].fillna("")

# Step 3: Left join merged_table with levels_of_work_df on ["Class Code", "Class Title"]
merged_table = merged_table.merge(
    levels_of_work_df,
    on=["Class Code", "Class Title"],
    how="left"
)

# Fill empty values in levels_of_work_df["Levels Of Work"] with an empty string
if "Levels Of Work" in merged_table.columns:
    merged_table["Levels Of Work"] = merged_table["Levels Of Work"].fillna("")

# Step 4: Left join merged_table with maq_df on ["Class Code", "Class Title"]
merged_table = merged_table.merge(
    maq_df,
    on=["Class Code", "Class Title"],
    how="left"
)

# Fill empty values in maq_df["MAQ"] with an empty string
if "MAQ" in merged_table.columns:
    merged_table["MAQ"] = merged_table["MAQ"].fillna("")

# Display the final merged table
print(merged_table.head())
```
